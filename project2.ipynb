# Computer Vision Problem
 The train and test datasets have been uploaded. This dataset has 4 classes. Labels for
training data are provided, you have to find labels of test data. Feel free to use any Machine learning or Deep learning technique.

##Unpickling the given files
Python pickle module is used for serializing and de-serializing a Python object structure. Any object in Python can be pickled so that it can be saved on disk. What pickle does is that it “serializes” the object first before writing it to file. Pickling is a way to convert a python object (list, dict, etc.) into a character stream. The idea is that this character stream contains all the information necessary to reconstruct the object in another python script.

```
import pickle
```

```
infile = open("train_image.pkl",'rb') #open the file in read in binary mode
infile1 = open("train_label.pkl",'rb')
infile2 = open("test_image.pkl",'rb')
new_dict = pickle.load(infile) #pickle.load() returns a file in dictionary format
new_dict1 = pickle.load(infile1)
new_dict2 = pickle.load(infile2)
infile.close() #closing the file opened
infile1.close()
infile2.close()
```

##Converting into an numpy array

```
import pandas as pd
import numpy as np
```

```
#converting the dictionary into a Dataframe
df1=pd.DataFrame(new_dict)
df1.to_csv("2.csv",index=True)#view the files in csv format
df2=pd.DataFrame(new_dict1)
df2.to_csv("1.csv",index=True)
df3=pd.DataFrame(new_dict2)
df3.to_csv("3.csv",index=True)
df3
```

```
#converting the dataframe in a numpy array
train_images=df1.values.tolist()
train_images=np.array(train_images)
train_label=df2.values.tolist()
train_label=np.array(train_label)
test_images=df3.values.tolist()
test_images=np.array(test_images)
```

##Reducing the range of values to (0,1)

```
train_images=train_images / 255.0
test_images=test_images/255.0
train_label=train_label
```

```
#reshaping the values inorder to fit into theneural network
train_images=train_images.reshape(8000, 28, 28, 1)
test_images = test_images.reshape(2000, 28, 28, 1)
```

##Making a Convolution Neural Network
A convolution neural network performs better than any Deep neural network in terms of accuracy by the end of 20 epochs we get an accuracy > 97%

```
import tensorflow as tf
model = tf.keras.models.Sequential([
  #applying the convolution to the first layer
  tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(28, 28, 1)),
    #pooling decreses the size of the image by half
  tf.keras.layers.MaxPooling2D(2, 2),
    #applying convolution the second time
  tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
  tf.keras.layers.MaxPooling2D(2,2),
    #Flatten makes the image one dimentional
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(10,activation="softmax")
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_images, train_label, epochs=20)
```

```
#predicting the probability of the image belonging to each class
y=model.predict(test_images)
y
```

```
#viewing the result as a dataframe
df=pd.DataFrame(y)
df

```

```
#conveting the predicted output array into a numpy array 
k=np.array(y)
#now finding the class with the maximum probability and saving it as our result
p=np.argmax(k,axis=1)
#converting the new array into a dataframe
df=pd.DataFrame(p,columns=["class"])
```

```
df.to_csv("5.csv",index=True)#converting into a csv file. index=True gives the corresponding index to the values
```
